```python
# êµì°¨ë¡œë³„ ì‹¤ì‹œê°„ ë°ì´í„° (ë¦¬ìŠ¤íŠ¸ + íŠœí”Œ)
intersections = [
    ('ì œì¼CCì‚¬ê±°', 'red', 25, 12),      # ì´ë¦„, ì‹ í˜¸ìƒ‰, ë‚¨ì€ì‹œê°„, ëŒ€ê¸°ì°¨ëŸ‰
    ('í™”ì •ì²œì„œ', 'green', 15, 8),
    ('ì¢…ë¡œ3ê°€', 'yellow', 3, 15),
    ('ì„ì§€ë¡œì…êµ¬', 'red', 40, 5)
]

processed_intersections = set()
intersection_details = {
    'ì„œìš¸ì—­ì•': {'priority': 'high', 'camera_working': True},
    'ì‹œì²­ì•': {'priority': 'medium', 'camera_working': False},
    'ì¢…ë¡œ3ê°€': {'priority': 'high', 'camera_working': True},
    'ì„ì§€ë¡œì…êµ¬': {'priority': 'low', 'camera_working': True}
}

print("êµì°¨ë¡œ í˜„í™© ë¶„ì„:")
for intersection_data in intersections:  # forë¬¸ìœ¼ë¡œ ë¦¬ìŠ¤íŠ¸ ìˆœíšŒ
    name, signal, time_left, waiting_cars = intersection_data  # íŠœí”Œ ì–¸íŒ¨í‚¹

    # ë”•ì…”ë„ˆë¦¬ì—ì„œ ì¶”ê°€ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    details = intersection_details[name]
    priority = details['priority']
    camera_status = details['camera_working']

    print(f"\n{name}:")
    print(f"  ì‹ í˜¸: {signal} ({time_left}ì´ˆ)")
    print(f"  ëŒ€ê¸°ì°¨ëŸ‰: {waiting_cars}ëŒ€")
    print(f"  ìš°ì„ ìˆœìœ„: {priority}")
    print(f"  CCTV ì‘ë™: {'ì •ìƒ' if camera_status else 'ê³ ì¥'}")

    if waiting_cars >= 10 and priority == 'high':
        print(" ê¸´ê¸‰ì²˜ë¦¬ í•„ìš”!")
        processed_intersections.add(name)  # ì…‹ì— ì¶”ê°€
    elif signal == 'yellow' and time_left <= 5:
        print(" ì‹ í˜¸ë³€ê²½!")
    elif not camera_status:
        print(" CCTV ìˆ˜ë¦¬ í•„ìš”")

print(f"\nê¸´ê¸‰ì²˜ë¦¬ ëŒ€ìƒ: {processed_intersections}")
```

```python
# ì „ê¸°ì°¨ ì •ë³´ (ë¦¬ìŠ¤íŠ¸ + íŠœí”Œ)
electric_vehicles = [
    ('EV001', 25, (127.1, 37.5), 'sedan'),     # ID, ë°°í„°ë¦¬%, ìœ„ì¹˜, ì°¨ì¢…
    ('EV002', 15, (127.2, 37.6), 'suv'),
    ('EV003', 80, (127.3, 37.4), 'truck'),
    ('EV004', 5, (127.0, 37.3), 'sedan'),
    ('EV005', 45, (127.15, 37.55), 'bus')
]

charging_stations = {
    'ê°•ë‚¨ì¶©ì „ì†Œ': {
        'location': (127.1, 37.5),
        'total_slots': 8,
        'available_slots': 3,
        'fast_charge': True
    },
    'ì„œì´ˆì¶©ì „ì†Œ': {
        'location': (127.2, 37.6),
        'total_slots': 5,
        'available_slots': 0,
        'fast_charge': False
    },
    'ì¢…ë¡œì¶©ì „ì†Œ': {
        'location': (127.0, 37.3),
        'total_slots': 6,
        'available_slots': 2,
        'fast_charge': True
    }
}

urgent_vehicles = set()
charging_recommendations = {}  # ë”•ì…”ë„ˆë¦¬

print("ì „ê¸°ì°¨ ë°°í„°ë¦¬ ìƒíƒœ:")
for vehicle_data in electric_vehicles:  # forë¬¸ìœ¼ë¡œ ì°¨ëŸ‰ ë°ì´í„° ìˆœíšŒ
    vehicle_id, battery, location, car_type = vehicle_data  # íŠœí”Œ ì–¸íŒ¨í‚¹
    lon, lat = location  # ìœ„ì¹˜ íŠœí”Œ ì–¸íŒ¨í‚¹

    print(f"{vehicle_id} ({car_type}): ë°°í„°ë¦¬ {battery}%, ìœ„ì¹˜ ({lon}, {lat})")

    # ifë¬¸ìœ¼ë¡œ ê¸´ê¸‰ë„ íŒë‹¨
    if battery <= 20:
        urgent_vehicles.add(vehicle_id)  # ì…‹ì— ê¸´ê¸‰ ì°¨ëŸ‰ ì¶”ê°€
        print(f" ê¸´ê¸‰ ì¶©ì „ í•„ìš”!")

        best_station = None
        min_distance = float('inf')

        for station_name, station_info in charging_stations.items():  # ë”•ì…”ë„ˆë¦¬ ìˆœíšŒ
            if station_info['available_slots'] > 0:  # ì´ìš© ê°€ëŠ¥í•œ ì¶©ì „ì†Œë§Œ
                station_lon, station_lat = station_info['location']  # íŠœí”Œ ì–¸íŒ¨í‚¹
                # ê°„ë‹¨í•œ ê±°ë¦¬ ê³„ì‚° (ì‹¤ì œë¡œëŠ” ë” ë³µì¡í•œ ê³„ì‚° ì‚¬ìš©)
                distance = abs(lon - station_lon) + abs(lat - station_lat)

                if distance < min_distance:
                    min_distance = distance
                    best_station = station_name

        if best_station:
            charging_recommendations[vehicle_id] = best_station  # ë”•ì…”ë„ˆë¦¬ì— ì¶”ì²œ ì €ì¥
            print(f" ì¶”ì²œ ì¶©ì „ì†Œ: {best_station}")
    elif battery <= 50:
        print(f" ì¶©ì „ ê¶Œì¥")
    else:
        print(f" ë°°í„°ë¦¬ ì¶©ë¶„")

print(f"\nê¸´ê¸‰ ì¶©ì „ í•„ìš” ì°¨ëŸ‰: {urgent_vehicles}")
print("ì¶©ì „ì†Œ ì¶”ì²œ:")
for vehicle, station in charging_recommendations.items():  # ë”•ì…”ë„ˆë¦¬ ìˆœíšŒ
    print(f"  {vehicle} â†’ {station}")
```

```python
# ë„ë¡œ êµ¬ê°„ë³„ ìœ„í—˜ ìš”ì†Œ (ë¦¬ìŠ¤íŠ¸ + íŠœí”Œ)
road_incidents = [
    ('ê°•ë‚¨ëŒ€ë¡œ_1êµ¬ê°„', 'accident', (127.1, 37.5), 'ê²½ë¯¸'),     # êµ¬ê°„ëª…, ì‚¬ê±´ì¢…ë¥˜, ìœ„ì¹˜, ì‹¬ê°ë„
    ('í…Œí—¤ë€ë¡œ_2êµ¬ê°„', 'construction', (127.2, 37.6), 'ë³´í†µ'),
    ('ì˜¬ë¦¼í”½ëŒ€ë¡œ_3êµ¬ê°„', 'heavy_rain', (127.3, 37.4), 'ì‹¬ê°'),
    ('í•œê°•ëŒ€ë¡œ_1êµ¬ê°„', 'fog', (127.0, 37.3), 'ë³´í†µ'),
    ('ê°•ë‚¨ëŒ€ë¡œ_2êµ¬ê°„', 'accident', (127.15, 37.55), 'ì‹¬ê°')
]

# ìœ„í—˜ êµ¬ê°„ ì¶”ì  ì…‹
dangerous_sections = set()
blocked_roads = set()

# êµ¬ê°„ë³„ í‰ìƒì‹œ êµí†µëŸ‰ ë”•ì…”ë„ˆë¦¬
normal_traffic = {
    'ê°•ë‚¨ëŒ€ë¡œ_1êµ¬ê°„': 150,
    'í…Œí—¤ë€ë¡œ_2êµ¬ê°„': 120,
    'ì˜¬ë¦¼í”½ëŒ€ë¡œ_3êµ¬ê°„': 200,
    'í•œê°•ëŒ€ë¡œ_1êµ¬ê°„': 100,
    'ê°•ë‚¨ëŒ€ë¡œ_2êµ¬ê°„': 180
}

# ìš°íšŒ ê²½ë¡œ ë”•ì…”ë„ˆë¦¬
alternative_routes = {
    'ê°•ë‚¨ëŒ€ë¡œ_1êµ¬ê°„': ['í…Œí—¤ë€ë¡œ_2êµ¬ê°„', 'í•œê°•ëŒ€ë¡œ_1êµ¬ê°„'],
    'í…Œí—¤ë€ë¡œ_2êµ¬ê°„': ['ì˜¬ë¦¼í”½ëŒ€ë¡œ_3êµ¬ê°„'],
    'ì˜¬ë¦¼í”½ëŒ€ë¡œ_3êµ¬ê°„': ['í•œê°•ëŒ€ë¡œ_1êµ¬ê°„'],
    'í•œê°•ëŒ€ë¡œ_1êµ¬ê°„': ['ê°•ë‚¨ëŒ€ë¡œ_2êµ¬ê°„'],
    'ê°•ë‚¨ëŒ€ë¡œ_2êµ¬ê°„': ['í…Œí—¤ë€ë¡œ_2êµ¬ê°„']
}

print("ë„ë¡œ ì•ˆì „ ìƒí™© ë¶„ì„:")
for incident_data in road_incidents:  # forë¬¸ìœ¼ë¡œ ì‚¬ê±´ ë°ì´í„° ìˆœíšŒ
    section, incident_type, location, severity = incident_data  # íŠœí”Œ ì–¸íŒ¨í‚¹
    lon, lat = location  # ìœ„ì¹˜ íŠœí”Œ ì–¸íŒ¨í‚¹

    print(f"\n{section}:")
    print(f"  ì‚¬ê±´: {incident_type}")
    print(f"  ìœ„ì¹˜: ({lon}, {lat})")
    print(f"  ì‹¬ê°ë„: {severity}")

    # ifë¬¸ìœ¼ë¡œ ëŒ€ì‘ ë°©ì•ˆ ê²°ì •
    if severity == 'ì‹¬ê°':
        dangerous_sections.add(section)  # ì…‹ì— ìœ„í—˜ êµ¬ê°„ ì¶”ê°€
        if incident_type in ['accident', 'construction']:
            blocked_roads.add(section)  # ì…‹ì— ì°¨ë‹¨ ë„ë¡œ ì¶”ê°€
            print(f"  ë„ë¡œ ì°¨ë‹¨ í•„ìš”!")
        else:
            print(f"  ì†ë„ ì œí•œ í•„ìš”!")
    elif severity == 'ë³´í†µ':
        print(f" ì£¼ì˜ ìš´í–‰ í•„ìš”")
    else:
        print(f" ê²½ë¯¸í•œ ì£¼ì˜")

print(f"\nìœ„í—˜ êµ¬ê°„: {dangerous_sections}")
print(f"ì°¨ë‹¨ í•„ìš” ë„ë¡œ: {blocked_roads}")

# ìš°íšŒ ê²½ë¡œ ì•ˆë‚´
print("\nìš°íšŒ ê²½ë¡œ ì•ˆë‚´:")
for blocked_section in blocked_roads:  # ì…‹ ìˆœíšŒ
    if blocked_section in alternative_routes:  # ë”•ì…”ë„ˆë¦¬ì—ì„œ í™•ì¸
        alternatives = alternative_routes[blocked_section]  # ë¦¬ìŠ¤íŠ¸ ê°€ì ¸ì˜¤ê¸°
        print(f"{blocked_section} ìš°íšŒ ê²½ë¡œ:")
        for alt_route in alternatives:  # ë¦¬ìŠ¤íŠ¸ ìˆœíšŒ
            # ìš°íšŒ ê²½ë¡œë„ ì•ˆì „í•œì§€ í™•ì¸
            if alt_route not in dangerous_sections:  # ì…‹ì—ì„œ í™•ì¸
                traffic_volume = normal_traffic[alt_route]  # ë”•ì…”ë„ˆë¦¬ì—ì„œ ê°’ ê°€ì ¸ì˜¤ê¸°
                print(f"  {alt_route} (í‰ì†Œ êµí†µëŸ‰: {traffic_volume}ëŒ€/ì‹œê°„)")
            else:
                print(f" {alt_route} (ìœ„í—˜êµ¬ê°„ìœ¼ë¡œ ë¹„ì¶”ì²œ)")
```

```python
# ì°¨ëŸ‰ ì£¼ì°¨ ìš”ì²­ (ë¦¬ìŠ¤íŠ¸ + íŠœí”Œ)
parking_requests = [
    ('car_001', 'compact', 2, 'ì¼ë°˜'),      # ì°¨ëŸ‰ID, í¬ê¸°, ì˜ˆìƒì£¼ì°¨ì‹œê°„, ìš°ì„ ìˆœìœ„
    ('ambulance_01', 'standard', 1, 'ì‘ê¸‰'),
    ('truck_001', 'large', 4, 'ì¼ë°˜'),
    ('car_002', 'compact', 3, 'ì¼ë°˜'),
    ('police_01', 'standard', 2, 'ê¸´ê¸‰')
]

# ì£¼ì°¨ êµ¬ì—­ ì •ë³´ ë”•ì…”ë„ˆë¦¬
parking_zones = {
    'Aêµ¬ì—­': {
        'car_type': 'compact',  #ì°¨ëŸ‰ ìœ í˜•
        'total_spaces': 20,      # ì´ ì£¼ì°¨ê³µê°„
        'occupied_spaces': 18,   # ì‚¬ìš©ì¤‘ì¸ ê³µê°„
        'hourly_rate': 1000     # ì‹œê°„ë‹¤ ìš”ê¸ˆ
    },
    'Bêµ¬ì—­': {
        'car_type': 'standard',
        'total_spaces': 15,
        'occupied_spaces': 10,
        'hourly_rate': 1500
    },
    'Cêµ¬ì—­': {
        'car_type': 'large',
        'total_spaces': 10,
        'occupied_spaces': 8,
        'hourly_rate': 2000
    },
    'VIPêµ¬ì—­': {
        'car_type': 'standard',#ì°¨ëŸ‰ ìœ í˜•
        'total_spaces': 5, # ì´ ì£¼ì°¨ê³µê°„
        'occupied_spaces': 2, # ì‚¬ìš©ì¤‘ì¸ ê³µê°„
        'hourly_rate': 3000 # ì‹œê°„ë‹¤ ìš”ê¸ˆ
    } # Missing closing brace was here
}

# ì²˜ë¦¬ëœ ìš”ì²­ê³¼ ê±°ë¶€ëœ ìš”ì²­ ì¶”ì 
processed_requests = set()
rejected_requests = set()
priority_queue = []  # ìš°ì„ ìˆœìœ„ ìš”ì²­ ë¦¬ìŠ¤íŠ¸

print("ì£¼ì°¨ ìš”ì²­ ì²˜ë¦¬:")
# ìš°ì„ ìˆœìœ„ë³„ë¡œ ë¨¼ì € ë¶„ë¥˜
for request_data in parking_requests:  # forë¬¸ìœ¼ë¡œ ìš”ì²­ ìˆœíšŒ
    vehicle_id, car_size, duration, priority = request_data  # íŠœí”Œ ì–¸íŒ¨í‚¹

    if priority in ['ì‘ê¸‰', 'ê¸´ê¸‰']:  # ifë¬¸ìœ¼ë¡œ ìš°ì„ ìˆœìœ„ íŒë‹¨
        priority_queue.append(request_data)  # ë¦¬ìŠ¤íŠ¸ì— ìš°ì„  ìš”ì²­ ì¶”ê°€

print("ìš°ì„ ì²˜ë¦¬ ìš”ì²­:")
for priority_request in priority_queue:  # ìš°ì„ ìˆœìœ„ ìš”ì²­ ë¨¼ì € ì²˜ë¦¬
    vehicle_id, car_size, duration, priority = priority_request  # íŠœí”Œ ì–¸íŒ¨í‚¹
    print(f" {vehicle_id} ({priority}): {car_size} ì°¨ëŸ‰, {duration}ì‹œê°„")

    # ì í•©í•œ ì£¼ì°¨êµ¬ì—­ ì°¾ê¸°
    assigned = False
    for zone_name, zone_info in parking_zones.items():  # ë”•ì…”ë„ˆë¦¬ ìˆœíšŒ
        available = zone_info['total_spaces'] - zone_info['occupied_spaces']

        # ì°¨ëŸ‰ í¬ê¸° í˜¸í™˜ì„± í™•ì¸ (ifë¬¸)
        if ((car_size == 'compact' and zone_info['car_type'] in ['compact', 'standard', 'large']) or
            (car_size == 'standard' and zone_info['car_type'] in ['standard', 'large']) or
            (car_size == 'large' and zone_info['car_type'] == 'large')):

            if available > 0:  # ë¹ˆìë¦¬ ìˆìœ¼ë©´
                print(f" {zone_name} ë°°ì •")
                zone_info['occupied_spaces'] += 1  # ë”•ì…”ë„ˆë¦¬ ê°’ ì—…ë°ì´íŠ¸
                processed_requests.add(vehicle_id)  # ì…‹ì— ì²˜ë¦¬ì™„ë£Œ ì¶”ê°€
                assigned = True
                break

    if not assigned:
        print(f" ì£¼ì°¨ ë¶ˆê°€")
        rejected_requests.add(vehicle_id)  # ì…‹ì— ê±°ë¶€ ì¶”ê°€

# ì¼ë°˜ ìš”ì²­ ì²˜ë¦¬
print("\nì¼ë°˜ ìš”ì²­:")
for request_data in parking_requests:  # forë¬¸ìœ¼ë¡œ ëª¨ë“  ìš”ì²­ ì¬ìˆœíšŒ
    vehicle_id, car_size, duration, priority = request_data  # íŠœí”Œ ì–¸íŒ¨í‚¹

    # ì´ë¯¸ ì²˜ë¦¬ëœ ìš°ì„ ìˆœìœ„ ìš”ì²­ì€ ê±´ë„ˆë›°ê¸°
    if priority in ['ì‘ê¸‰', 'ê¸´ê¸‰']:
        continue

    print(f"{vehicle_id}: {car_size} ì°¨ëŸ‰, {duration}ì‹œê°„")

    # ìµœì  êµ¬ì—­ ì°¾ê¸° (ìš”ê¸ˆê³¼ ê°€ìš©ì„± ê³ ë ¤)
    best_zone = None
    min_cost = float('inf')

    for zone_name, zone_info in parking_zones.items():  # ë”•ì…”ë„ˆë¦¬ ìˆœíšŒ
        available = zone_info['total_spaces'] - zone_info['occupied_spaces']

        # ì°¨ëŸ‰ í¬ê¸° í˜¸í™˜ì„±ê³¼ ë¹ˆìë¦¬ í™•ì¸
        if available > 0:
            if ((car_size == 'compact' and zone_info['car_type'] in ['compact', 'standard']) or
                (car_size == 'standard' and zone_info['car_type'] == 'standard') or
                (car_size == 'large' and zone_info['car_type'] == 'large')):

                cost = zone_info['hourly_rate'] * duration
                if cost < min_cost:  # ê°€ì¥ ì €ë ´í•œ êµ¬ì—­ ì„ íƒ
                    min_cost = cost
                    best_zone = zone_name

    if best_zone:
        parking_zones[best_zone]['occupied_spaces'] += 1  # ë”•ì…”ë„ˆë¦¬ ì—…ë°ì´íŠ¸
        processed_requests.add(vehicle_id)  # ì…‹ì— ì¶”ê°€
        print(f"  {best_zone} ë°°ì • (ìš”ê¸ˆ: {min_cost}ì›)")
    else:
        rejected_requests.add(vehicle_id)  # ì…‹ì— ì¶”ê°€
        print(f" ì£¼ì°¨ ë¶ˆê°€")

print(f"\nì²˜ë¦¬ì™„ë£Œ: {processed_requests}")
print(f"ê±°ë¶€ë¨: {rejected_requests}")

# í˜„ì¬ ì£¼ì°¨ì¥ ìƒí™©
print("\ní˜„ì¬ ì£¼ì°¨ì¥ í˜„í™©:")
for zone_name, zone_info in parking_zones.items():  # ë”•ì…”ë„ˆë¦¬ ìˆœíšŒ
    available = zone_info['total_spaces'] - zone_info['occupied_spaces']
    utilization = (zone_info['occupied_spaces'] / zone_info['total_spaces']) * 100
    print(f"{zone_name}: {available}/{zone_info['total_spaces']} ê°€ëŠ¥ (ì‚¬ìš©ë¥  {utilization:.1f}%)")
```
