 # 종합 자율주행 과제 - 튜플, 리스트, 셋, 딕셔너리, if문, for문 모두 활용


```python
# 교차로별 실시간 데이터 (리스트 + 튜플)
intersections = [
    ('제일CC사거', 'red', 25, 12),      # 이름, 신호색, 남은시간, 대기차량
    ('화정천서', 'green', 15, 8),
    ('종로3가', 'yellow', 3, 15),
    ('을지로입구', 'red', 40, 5)
]

processed_intersections = set()
intersection_details = {
    '서울역앞': {'priority': 'high', 'camera_working': True},
    '시청앞': {'priority': 'medium', 'camera_working': False},
    '종로3가': {'priority': 'high', 'camera_working': True},
    '을지로입구': {'priority': 'low', 'camera_working': True}
}

print("교차로 현황 분석:")
for intersection_data in intersections:  # for문으로 리스트 순회
    name, signal, time_left, waiting_cars = intersection_data  # 튜플 언패킹

    # 딕셔너리에서 추가 정보 가져오기
    details = intersection_details[name]
    priority = details['priority']
    camera_status = details['camera_working']

    print(f"\n{name}:")
    print(f"  신호: {signal} ({time_left}초)")
    print(f"  대기차량: {waiting_cars}대")
    print(f"  우선순위: {priority}")
    print(f"  CCTV 작동: {'정상' if camera_status else '고장'}")

    if waiting_cars >= 10 and priority == 'high':
        print(" 긴급처리 필요!")
        processed_intersections.add(name)  # 셋에 추가
    elif signal == 'yellow' and time_left <= 5:
        print(" 신호변경!")
    elif not camera_status:
        print(" CCTV 수리 필요")

print(f"\n긴급처리 대상: {processed_intersections}")
```

전기차 충전소 최적화 시스템

```python
# 전기차 정보 (리스트 + 튜플)
electric_vehicles = [
    ('EV001', 25, (127.1, 37.5), 'sedan'),     # ID, 배터리%, 위치, 차종
    ('EV002', 15, (127.2, 37.6), 'suv'),
    ('EV003', 80, (127.3, 37.4), 'truck'),
    ('EV004', 5, (127.0, 37.3), 'sedan'),
    ('EV005', 45, (127.15, 37.55), 'bus')
]

charging_stations = {
    '강남충전소': {
        'location': (127.1, 37.5),
        'total_slots': 8,
        'available_slots': 3,
        'fast_charge': True
    },
    '서초충전소': {
        'location': (127.2, 37.6),
        'total_slots': 5,
        'available_slots': 0,
        'fast_charge': False
    },
    '종로충전소': {
        'location': (127.0, 37.3),
        'total_slots': 6,
        'available_slots': 2,
        'fast_charge': True
    }
}

urgent_vehicles = set()
charging_recommendations = {}  # 딕셔너리

print("전기차 배터리 상태:")
for vehicle_data in electric_vehicles:  # for문으로 차량 데이터 순회
    vehicle_id, battery, location, car_type = vehicle_data  # 튜플 언패킹
    lon, lat = location  # 위치 튜플 언패킹

    print(f"{vehicle_id} ({car_type}): 배터리 {battery}%, 위치 ({lon}, {lat})")

    # if문으로 긴급도 판단
    if battery <= 20:
        urgent_vehicles.add(vehicle_id)  # 셋에 긴급 차량 추가
        print(f" 긴급 충전 필요!")

        best_station = None
        min_distance = float('inf')

        for station_name, station_info in charging_stations.items():  # 딕셔너리 순회
            if station_info['available_slots'] > 0:  # 이용 가능한 충전소만
                station_lon, station_lat = station_info['location']  # 튜플 언패킹
                # 간단한 거리 계산 (실제로는 더 복잡한 계산 사용)
                distance = abs(lon - station_lon) + abs(lat - station_lat)

                if distance < min_distance:
                    min_distance = distance
                    best_station = station_name

        if best_station:
            charging_recommendations[vehicle_id] = best_station  # 딕셔너리에 추천 저장
            print(f" 추천 충전소: {best_station}")
    elif battery <= 50:
        print(f" 충전 권장")
    else:
        print(f" 배터리 충분")

print(f"\n긴급 충전 필요 차량: {urgent_vehicles}")
print("충전소 추천:")
for vehicle, station in charging_recommendations.items():  # 딕셔너리 순회
    print(f"  {vehicle} → {station}")
```

도로 안전 모니터링 시스템

```python
# 도로 구간별 위험 요소 (리스트 + 튜플)
road_incidents = [
    ('강남대로_1구간', 'accident', (127.1, 37.5), '경미'),     # 구간명, 사건종류, 위치, 심각도
    ('테헤란로_2구간', 'construction', (127.2, 37.6), '보통'),
    ('올림픽대로_3구간', 'heavy_rain', (127.3, 37.4), '심각'),
    ('한강대로_1구간', 'fog', (127.0, 37.3), '보통'),
    ('강남대로_2구간', 'accident', (127.15, 37.55), '심각')
]

# 위험 구간 추적 셋
dangerous_sections = set()
blocked_roads = set()

# 구간별 평상시 교통량 딕셔너리
normal_traffic = {
    '강남대로_1구간': 150,
    '테헤란로_2구간': 120,
    '올림픽대로_3구간': 200,
    '한강대로_1구간': 100,
    '강남대로_2구간': 180
}

# 우회 경로 딕셔너리
alternative_routes = {
    '강남대로_1구간': ['테헤란로_2구간', '한강대로_1구간'],
    '테헤란로_2구간': ['올림픽대로_3구간'],
    '올림픽대로_3구간': ['한강대로_1구간'],
    '한강대로_1구간': ['강남대로_2구간'],
    '강남대로_2구간': ['테헤란로_2구간']
}

print("도로 안전 상황 분석:")
for incident_data in road_incidents:  # for문으로 사건 데이터 순회
    section, incident_type, location, severity = incident_data  # 튜플 언패킹
    lon, lat = location  # 위치 튜플 언패킹

    print(f"\n{section}:")
    print(f"  사건: {incident_type}")
    print(f"  위치: ({lon}, {lat})")
    print(f"  심각도: {severity}")

    # if문으로 대응 방안 결정
    if severity == '심각':
        dangerous_sections.add(section)  # 셋에 위험 구간 추가
        if incident_type in ['accident', 'construction']:
            blocked_roads.add(section)  # 셋에 차단 도로 추가
            print(f"  도로 차단 필요!")
        else:
            print(f"  속도 제한 필요!")
    elif severity == '보통':
        print(f" 주의 운행 필요")
    else:
        print(f" 경미한 주의")

print(f"\n위험 구간: {dangerous_sections}")
print(f"차단 필요 도로: {blocked_roads}")

# 우회 경로 안내
print("\n우회 경로 안내:")
for blocked_section in blocked_roads:  # 셋 순회
    if blocked_section in alternative_routes:  # 딕셔너리에서 확인
        alternatives = alternative_routes[blocked_section]  # 리스트 가져오기
        print(f"{blocked_section} 우회 경로:")
        for alt_route in alternatives:  # 리스트 순회
            # 우회 경로도 안전한지 확인
            if alt_route not in dangerous_sections:  # 셋에서 확인
                traffic_volume = normal_traffic[alt_route]  # 딕셔너리에서 값 가져오기
                print(f"  {alt_route} (평소 교통량: {traffic_volume}대/시간)")
            else:
                print(f" {alt_route} (위험구간으로 비추천)")
```

 스마트 주차 관리 시스템

```python
# 차량 주차 요청 (리스트 + 튜플)
parking_requests = [
    ('car_001', 'compact', 2, '일반'),      # 차량ID, 크기, 예상주차시간, 우선순위
    ('ambulance_01', 'standard', 1, '응급'),
    ('truck_001', 'large', 4, '일반'),
    ('car_002', 'compact', 3, '일반'),
    ('police_01', 'standard', 2, '긴급')
]

# 주차 구역 정보 딕셔너리
parking_zones = {
    'A구역': {
        'car_type': 'compact',  #차량 유형
        'total_spaces': 20,      # 총 주차공간
        'occupied_spaces': 18,   # 사용중인 공간
        'hourly_rate': 1000     # 시간다 요금
    },
    'B구역': {
        'car_type': 'standard',
        'total_spaces': 15,
        'occupied_spaces': 10,
        'hourly_rate': 1500
    },
    'C구역': {
        'car_type': 'large',
        'total_spaces': 10,
        'occupied_spaces': 8,
        'hourly_rate': 2000
    },
    'VIP구역': {
        'car_type': 'standard',#차량 유형
        'total_spaces': 5, # 총 주차공간
        'occupied_spaces': 2, # 사용중인 공간
        'hourly_rate': 3000 # 시간다 요금
    } # Missing closing brace was here
}

# 처리된 요청과 거부된 요청 추적
processed_requests = set()
rejected_requests = set()
priority_queue = []  # 우선순위 요청 리스트

print("주차 요청 처리:")
# 우선순위별로 먼저 분류
for request_data in parking_requests:  # for문으로 요청 순회
    vehicle_id, car_size, duration, priority = request_data  # 튜플 언패킹

    if priority in ['응급', '긴급']:  # if문으로 우선순위 판단
        priority_queue.append(request_data)  # 리스트에 우선 요청 추가

print("우선처리 요청:")
for priority_request in priority_queue:  # 우선순위 요청 먼저 처리
    vehicle_id, car_size, duration, priority = priority_request  # 튜플 언패킹
    print(f" {vehicle_id} ({priority}): {car_size} 차량, {duration}시간")

    # 적합한 주차구역 찾기
    assigned = False
    for zone_name, zone_info in parking_zones.items():  # 딕셔너리 순회
        available = zone_info['total_spaces'] - zone_info['occupied_spaces']

        # 차량 크기 호환성 확인 (if문)
        if ((car_size == 'compact' and zone_info['car_type'] in ['compact', 'standard', 'large']) or
            (car_size == 'standard' and zone_info['car_type'] in ['standard', 'large']) or
            (car_size == 'large' and zone_info['car_type'] == 'large')):

            if available > 0:  # 빈자리 있으면
                print(f" {zone_name} 배정")
                zone_info['occupied_spaces'] += 1  # 딕셔너리 값 업데이트
                processed_requests.add(vehicle_id)  # 셋에 처리완료 추가
                assigned = True
                break

    if not assigned:
        print(f" 주차 불가")
        rejected_requests.add(vehicle_id)  # 셋에 거부 추가

# 일반 요청 처리
print("\n일반 요청:")
for request_data in parking_requests:  # for문으로 모든 요청 재순회
    vehicle_id, car_size, duration, priority = request_data  # 튜플 언패킹

    # 이미 처리된 우선순위 요청은 건너뛰기
    if priority in ['응급', '긴급']:
        continue

    print(f"{vehicle_id}: {car_size} 차량, {duration}시간")

    # 최적 구역 찾기 (요금과 가용성 고려)
    best_zone = None
    min_cost = float('inf')

    for zone_name, zone_info in parking_zones.items():  # 딕셔너리 순회
        available = zone_info['total_spaces'] - zone_info['occupied_spaces']

        # 차량 크기 호환성과 빈자리 확인
        if available > 0:
            if ((car_size == 'compact' and zone_info['car_type'] in ['compact', 'standard']) or
                (car_size == 'standard' and zone_info['car_type'] == 'standard') or
                (car_size == 'large' and zone_info['car_type'] == 'large')):

                cost = zone_info['hourly_rate'] * duration
                if cost < min_cost:  # 가장 저렴한 구역 선택
                    min_cost = cost
                    best_zone = zone_name

    if best_zone:
        parking_zones[best_zone]['occupied_spaces'] += 1  # 딕셔너리 업데이트
        processed_requests.add(vehicle_id)  # 셋에 추가
        print(f"  {best_zone} 배정 (요금: {min_cost}원)")
    else:
        rejected_requests.add(vehicle_id)  # 셋에 추가
        print(f" 주차 불가")

print(f"\n처리완료: {processed_requests}")
print(f"거부됨: {rejected_requests}")

# 현재 주차장 상황
print("\n현재 주차장 현황:")
for zone_name, zone_info in parking_zones.items():  # 딕셔너리 순회
    available = zone_info['total_spaces'] - zone_info['occupied_spaces']
    utilization = (zone_info['occupied_spaces'] / zone_info['total_spaces']) * 100
    print(f"{zone_name}: {available}/{zone_info['total_spaces']} 가능 (사용률 {utilization:.1f}%)")
```
